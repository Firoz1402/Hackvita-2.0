```python
import cv2
import numpy as np

def find_image_in_video(image_path, video_path, threshold=0.7):
    # Load the reference image
    reference_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Load the video
    video_capture = cv2.VideoCapture(video_path)
    fps = video_capture.get(cv2.CAP_PROP_FPS)

    # Initialize variables
    found = False
    frame_count = 0
    timestamps = []

    # Loop through the video frames
    while True:
        ret, frame = video_capture.read()
        if not ret:
            break

        # Convert frame to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Try to find the reference image within the frame
        result = cv2.matchTemplate(gray_frame, reference_image, cv2.TM_CCOEFF_NORMED)
        min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

        if max_val > threshold:
            found = True
            timestamps.append(frame_count / fps)

        frame_count += 1

    video_capture.release()
    return timestamps if found else None

# Example usage
image_path = 'input_images/test2.jpg'
video_path = 'input_video/test_video.mp4'
timestamps = find_image_in_video(image_path, video_path)

if timestamps:
    print("Image found in video at timestamps:", timestamps)
else:
    print("Image not found in video.")
```

```python
from flask import Flask, jsonify, request
import cv2
import numpy as np

app = Flask(__name__)

@app.route('/detectimage', methods=['POST'])
def detect_image():
    # Load the reference image
    ref_img = cv2.imread(request.files['image'].read())

    # Load the video
    video = cv2.VideoCapture(request.files['video'].read())

    # Initialize arrays for storing the timestamps
    timestamps = []

    # Define the function for detecting the image in the video
    def detect(img, ref_img):
        # Convert the images to grayscale
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        ref_gray = cv2.cvtColor(ref_img, cv2.COLOR_BGR2GRAY)

        # Resize the reference image for faster matching
        resized_ref_gray = cv2.resize(ref_gray, (50,50), interpolation=cv2.INTER_AREA)

        # Compute the hash value for the reference image
        hash_value = np.sum(np.abs(np.diff(resized_ref_gray)))

        # Initialize the similarity threshold
        threshold = 0.5

        # Iterate through the video frames
        for frame in range(int(video.get(7))) :
            # Read the video frame
            video.set(cv2.CAP_PROP_POS_FRAMES, frame)
            ret, img_gray = video.read()

            # Resize the current frame for faster matching
            resized_img_gray = cv2.resize(img_gray, (50,50), interpolation=cv2.INTER_AREA)

            # Compute the hash value for the current frame
            frame_hash_value = np.sum(np.abs(np.diff(resized_img_gray)))
# Check if the hash value is similar to the reference image
            if abs(frame_hash_value - hash_value) < threshold:
                # Compute the similarity score
                score = cv2.matchTemplate(img_gray, ref_gray, cv2.TM_CCOEFF_NORMED)

                # Find the index of the maximum similarity score
                max_val = np.max(score)

                # Check if the similarity score is above a threshold
                if max_val > 0.5:
                    # Append the current timestamp to the array
                    timestamps.append(video.get(cv2.CAP_PROP_POS_MSEC))

    # Detect the image in the video
    detect(video.read(), ref_img)

    # Release the video
    video.release()

    # Return the timestamps as JSON
    return jsonify(timestamps)

if __name__ == '__main__':
    app.run(debug=True)
```
